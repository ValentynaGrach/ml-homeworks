4. Вставити 20 значень np.nan на випадкові позиції в масиві (при використанні звичайного рандому можуть накластись позиції, тому знайти рішення, яке гарантує 20 унікальних позицій)

```python
np.random.seed(100)

iris[np.unravel_index(np.random.choice(iris.size, size=20, replace=False), iris.shape)] = np.nan

nan_positions = np.where(np.isnan(iris))
print("Кількість значень np.nan:", len(nan_positions[0]))
print("Позиції значень np.nan:", list(zip(nan_positions[0], nan_positions[1])))
print("Чи є 20 np.nan?", len(nan_positions[0]) == 20)

print("Dataset form:", iris.shape)
print(iris[:5])
```

9. Розбити масив по вертикалі на 2 рівні частини (не використовувати абсолютні числа, мають бути два массиви по 4 колонки)

```python
iris1, iris2 = np.split(iris, 2, axis=0)

print("Dataset form:", iris1.shape)
print("Dataset form:", iris2.shape)
print("_________" * 10)
print(iris1[:5])
print("_________" * 10)
print(iris2[:5])
```

10. Відсортувати обидва массиви по 1-й колонці: 1-й за збільшенням, 2-й за зменшенням

```python
iris1 = iris1[np.argsort(iris1[:, 0])]
iris2 = iris2[np.argsort(iris2[:, 0])[::-1]]

print("Sorted upper:")
print(iris1)
print("_________" * 10)
print("Sorted lower:")
print(iris2)
```

11. Зібрати обидва массиви в одне ціле

13. Написати функцію, яка б множила всі значення в колонці, які менше середнього значения в цій колонці, на 2, і ділила інші значення на 4.

```python
def modify_column(column):
    mean_val = np.mean(column)
    column = np.where(column < mean_val, column * 2, column / 4)
    return column
```

14. Застосувати отриману функцію до 3-ї колонки

```python
iris3[:, 2] = modify_column(iris3[:, 2])
print(iris3)
```
