# üîπ –ö–†–û–ö 2: –Ü–º–ø–æ—Ä—Ç –±—ñ–±–ª—ñ–æ—Ç–µ–∫

```python
import numpy as np
import matplotlib.pyplot as plt
import gym
import imageio
from pyvirtualdisplay import Display
from base64 import b64encode
from IPython.display import HTML

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.callbacks import BaseCallback
```

# üîπ –ö–†–û–ö 3: –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞ LunarLander-v2

üåç –°—Ç–≤–æ—Ä—é—î–º–æ —Å–µ—Ä–µ–¥–æ–≤–∏—â–µ –¥–ª—è –Ω–∞–≤—á–∞–Ω–Ω—è –∞–≥–µ–Ω—Ç–∞ ‚Äì —Å–∏–º—É–ª—è—Ç–æ—Ä –ø–æ—Å–∞–¥–∫–∏ —Ä–∞–∫–µ—Ç–∏.

```python
env = make_vec_env("LunarLander-v3", n_envs=4)

print("üåç –°–µ—Ä–µ–¥–æ–≤–∏—â–µ LunarLander —Å—Ç–≤–æ—Ä–µ–Ω–æ.")
```

# üîπ –ö–†–û–ö 4: –ù–∞–≤—á–∞–Ω–Ω—è –∞–≥–µ–Ω—Ç–∞ –∑ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º PPO

üß† –ù–∞–≤—á–∞—î–º–æ –∞–≥–µ–Ω—Ç–∞ –∑ –Ω—É–ª—è –∫–µ—Ä—É–≤–∞—Ç–∏ —Ä–∞–∫–µ—Ç–æ—é –≤ —Å–µ—Ä–µ–¥–æ–≤–∏—â—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é PPO.

```python
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=200_000)
print("–ù–∞–≤—á–∞–Ω–Ω—è –∑–∞–≤–µ—Ä—à–µ–Ω–æ ‚úÖ")
```

üîç **–ê–Ω–∞–ª—ñ–∑ –Ω–∞–≤—á–µ–Ω–æ—ó –º–æ–¥–µ–ª—ñ**

‚Üí –º–æ–¥–µ–ª—å —Å—Ç–∞–±—ñ–ª—å–Ω–æ –Ω–∞–≤—á–∞—î—Ç—å—Å—è, –±–µ–∑ —Å—Ç—Ä–∏–±–∫—ñ–≤  

üìà **ep_rew_mean (—Å–µ—Ä–µ–¥–Ω—è –≤–∏–Ω–∞–≥–æ—Ä–æ–¥–∞)**:
- –ü–æ—á–∞—Ç–æ–∫: **-171**
- –ö—ñ–Ω–µ—Ü—å: **+98.6** ‚Üí üöÄ **–∑–Ω–∞—á–Ω–µ –ø–æ–∫—Ä–∞—â–µ–Ω–Ω—è!**

üìè **ep_len_mean** (—Å–µ—Ä–µ–¥–Ω—è –¥–æ–≤–∂–∏–Ω–∞ –µ–ø—ñ–∑–æ–¥—É):
- –ó—Ä–æ—Å–ª–∞ –∑ **88** –¥–æ **889** ‚Üí –∞–≥–µ–Ω—Ç –∂–∏–≤–µ –≤ —Ä–∞–∑–∏ –¥–æ–≤—à–µ

üìä **explained_variance**:
- –ó ~**-0.00** –¥–æ **0.815‚Äì0.94** ‚Üí –º–æ–¥–µ–ª—å **—á—É–¥–æ–≤–æ –ø–µ—Ä–µ–¥–±–∞—á–∞—î —Ü—ñ–ª—å–æ–≤—É —Ñ—É–Ω–∫—Ü—ñ—é**

‚úÖ **–í–∏—Å–Ω–æ–≤–æ–∫**:
üîπ –ú–æ–¥–µ–ª—å **—É—Å–ø—ñ—à–Ω–æ –Ω–∞–≤—á–∞—î—Ç—å—Å—è**  
üîπ –ê–≥–µ–Ω—Ç **–∑ –∫–æ–∂–Ω–æ—é —ñ—Ç–µ—Ä–∞—Ü—ñ—î—é –¥—ñ—î –∫—Ä–∞—â–µ**  
üîπ –í–∏–Ω–∞–≥–æ—Ä–æ–¥–∞ –∑—Ä–æ—Å–ª–∞ –∑ –Ω–µ–≥–∞—Ç–∏–≤–Ω–æ—ó –¥–æ —Å—Ç–∞–±—ñ–ª—å–Ω–æ **–ø–æ–∑–∏—Ç–∏–≤–Ω–æ—ó**  
üîπ –ù–∞–≤—á–∞–Ω–Ω—è –ø—Ä–æ–π—à–ª–æ **—à–≤–∏–¥–∫–æ —ñ –µ—Ñ–µ–∫—Ç–∏–≤–Ω–æ**

‚è≠ **–ù–∞—Å—Ç—É–ø–Ω–∏–π –∫—Ä–æ–∫** ‚Üí –≤–∏–≤–µ–¥–µ–º–æ –≤—ñ–¥–µ–æ, **—è–∫ –∞–≥–µ–Ω—Ç –≥—Ä–∞—î** üé•

# üîπ –ö–†–û–ö 5: –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –≥—Ä–∏ –∞–≥–µ–Ω—Ç–∞ (–≤—ñ–¥–µ–æ)

üé• –ó–∞–ø—É—Å–∫–∞—î–º–æ –æ–¥–∏–Ω –µ–ø—ñ–∑–æ–¥ –≥—Ä–∏, –∑–∞–ø–∏—Å—É—î–º–æ —ñ –ø–æ–∫–∞–∑—É—î–º–æ –≤—ñ–¥–µ–æ.

```python
# üß© –ü–∞—Ç—á –¥–ª—è —Å—É–º—ñ—Å–Ω–æ—Å—Ç—ñ –∑ NumPy >= 2.0
if not hasattr(np, 'bool8'):
    np.bool8 = np.bool_

display = Display(visible=0, size=(1400, 900))
display.start()

# üé• –í—ñ—Ä—Ç—É–∞–ª—å–Ω–∏–π –¥–∏—Å–ø–ª–µ–π
env = gym.make("LunarLander-v2", render_mode="rgb_array")
obs = env.reset()
video = []

# üéÆ –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Å–µ—Ä–µ–¥–æ–≤–∏—â–∞ LunarLander
for _ in range(500):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    frame = env.render()
    video.append(frame)
    if done:
        break

env.close()

# üßº –û–±—Ä–æ–±–∫–∞ –∫–∞–¥—Ä—ñ–≤
clean_video = []
for frame in video:
    if isinstance(frame, (list, tuple)):
        frame = frame[0]
    if len(frame.shape) == 2:
        frame = np.stack([frame] * 3, axis=-1)
    elif frame.shape[-1] not in [1, 3, 4]:
        continue
    clean_video.append(np.array(frame))

# üíæ –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –≤—ñ–¥–µ–æ
video_path = "/content/lunarlander.mp4"
imageio.mimsave(video_path, clean_video, fps=30)

# üëÅ –ü–æ–∫–∞–∑—É—î–º–æ –≤—ñ–¥–µ–æ
mp4 = open(video_path, 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML('<video width=600 controls autoplay loop><source src="' + data_url + '" type="video/mp4"></video>')
```
