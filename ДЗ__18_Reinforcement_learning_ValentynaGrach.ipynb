# 🔹 КРОК 2: Імпорт бібліотек

```python
import numpy as np
import matplotlib.pyplot as plt
import gym
import imageio
from pyvirtualdisplay import Display
from base64 import b64encode
from IPython.display import HTML

from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.callbacks import BaseCallback
```

# 🔹 КРОК 3: Створення середовища LunarLander-v2

🌍 Створюємо середовище для навчання агента – симулятор посадки ракети.

```python
env = make_vec_env("LunarLander-v3", n_envs=4)

print("🌍 Середовище LunarLander створено.")
```

# 🔹 КРОК 4: Навчання агента з алгоритмом PPO

🧠 Навчаємо агента з нуля керувати ракетою в середовищі за допомогою PPO.

```python
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=200_000)
print("Навчання завершено ✅")
```

# 🔹 КРОК 5: Візуалізація гри агента (відео)

🎥 Запускаємо один епізод гри, записуємо і показуємо відео.

```python
# 🧩 Патч для сумісності з NumPy >= 2.0
if not hasattr(np, 'bool8'):
    np.bool8 = np.bool_

display = Display(visible=0, size=(1400, 900))
display.start()

# 🎥 Віртуальний дисплей
env = gym.make("LunarLander-v2", render_mode="rgb_array")
obs = env.reset()
video = []

# 🎮 Створення середовища LunarLander
for _ in range(500):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, info = env.step(action)
    frame = env.render()
    video.append(frame)
    if done:
        break

env.close()

# 🧼 Обробка кадрів
clean_video = []
for frame in video:
    if isinstance(frame, (list, tuple)):
        frame = frame[0]
    if len(frame.shape) == 2:
        frame = np.stack([frame] * 3, axis=-1)
    elif frame.shape[-1] not in [1, 3, 4]:
        continue
    clean_video.append(np.array(frame))

# 💾 Зберігаємо відео
video_path = "/content/lunarlander.mp4"
imageio.mimsave(video_path, clean_video, fps=30)

# 👁 Показуємо відео
mp4 = open(video_path, 'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML('<video width=600 controls autoplay loop><source src="' + data_url + '" type="video/mp4"></video>')
```
